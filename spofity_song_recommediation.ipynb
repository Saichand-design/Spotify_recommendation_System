```python
#def recommend(song_name, top_n=10):

    # find song in data
    # compute similarity
    # return top_n results

from sklearn.metrics.pairwise import cosine_similarity

# Example: songs_df should have columns ['song_name', 'features']
# where 'features' is a vector (e.g., embedding, genre encoding, etc.)
def recommend(song_name, top_n=10, songs_df=None):
    """
    Recommend top_n similar songs to the given song_name.

    Parameters:
    - song_name (str): The name of the song to base recommendations on.
    - top_n (int): Number of recommendations to return (default=10).
    - songs_df (pd.DataFrame): DataFrame with 'song_name' and 'features'.

    Returns:
    - List of recommended song names.
    """
    if songs_df is None:
        raise ValueError("songs_df must be provided with song features.")

    # Check if song exists
    if song_name not in songs_df['song_name'].values:
        raise ValueError(f"'{song_name}' not found in dataset.")

    # Get features of the target song
    target_features = songs_df.loc[songs_df['song_name'] == song_name, 'features'].values[0].reshape(1, -1)

    # Compute similarity
    similarities = cosine_similarity(target_features, list(songs_df['features']))[0]

    # Create a DataFrame of results
    songs_df['similarity'] = similarities

    # Sort by similarity (excluding the song itself)
    recommendations = (
        songs_df[songs_df['song_name'] != song_name]
        .sort_values(by='similarity', ascending=False)
        .head(top_n)
    )

    return recommendations['song_name'].tolist
```
